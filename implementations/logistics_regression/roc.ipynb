```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve

# Generate a 2-class dataset
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)

# Split the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Dummy model with output always 0 (majority class)
dummy_model_prob = [0 for _ in range(len(y_test))]  # Majority class predictor (class 0)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Get predicted probabilities for class 1 (positive class)
model_pred_prob = model.predict_proba(X_test)[:, 1]

# Calculate AUC for the dummy model (always predicts 0)
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)

# Calculate AUC for the trained logistic regression model
model_auc = roc_auc_score(y_test, model_pred_prob)

# Print the AUC scores
print("Dummy Model AUC:", dummy_model_auc)
print("Trained Model AUC:", model_auc)

# Calculate ROC curve for the trained model
d_fpr, d_tpr, thresholds = roc_curve(y_test, dummy_model_prob)

# Calculate ROC curve for the trained model
fpr, tpr, thresholds = roc_curve(y_test, model_pred_prob)

# Plot ROC curve for the trained model
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='Logistic Regression (AUC = %0.2f)' % model_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random classifier line (diagonal)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Assuming you already have the following:
# y_test: true labels for the test set
# model_pred_prob: predicted probabilities for class 1 (positive class) from Logistic Regression model

# Dummy model (predicting always class 0)
dummy_model_prob = [0 for _ in range(len(y_test))]

# Get ROC curve for the dummy model (predicting always 0)
d_fpr, d_tpr, d_thresholds = roc_curve(y_test, dummy_model_prob)

# Get ROC curve for the logistic regression model
fpr, tpr, thresholds = roc_curve(y_test, model_pred_prob)

# AUC for the trained model
model_auc = roc_auc_score(y_test, model_pred_prob)

# Plot ROC curve for both the dummy model and the logistic regression model
fig = plt.figure(figsize=(20, 50))

# Plot dummy model ROC curve
plt.plot(d_fpr, d_tpr, linestyle='--', label='Dummy Model (AUC = %.2f)' % 0.5)  # AUC for dummy model is 0.5

# Plot logistic regression ROC curve
plt.plot(fpr, tpr, linestyle='-', label='Logistic Regression (AUC = %.2f)' % model_auc)

# Annotate threshold values on the ROC curve for logistic regression
ax = fig.add_subplot(111)
for xyz in zip(fpr, tpr, thresholds):
    ax.annotate(f'{np.round(xyz[2], 2)}', xy=(xyz[0], xyz[1]), textcoords='offset points', xytext=(0, 5), ha='center')

# Add labels and legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)

# Show the plot
plt.show()
```
